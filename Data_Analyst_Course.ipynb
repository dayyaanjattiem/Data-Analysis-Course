**Manipulating Pandas Dataframe**

Question: What are the 5 most popular baby names in 2014 in US?
- Slice out for 2014
- Sort rows in descending order by count
- Retrieve the first five rows


```python
us_babies ['Year']
```

```python
us_babies['Year'] == 2014
```

By utelising the Loc method, inside the .loc[*data you would like to access*]

```python
us_babies_2014  = us_babies.loc[us_babies['Year']==2014, :]
```

```python
us_babies_2014
```

```python
sorted_us_2014 = us_babies_2014.sort_values('Count', ascending=False)
```

```python
sorted_us_2014
```

Now we need to return the top 5 rows, by doing so we utelise the iLoc method,

```python
sorted_us_2014.iloc[0:5]
```

```python
listings = pd.read_csv("AB_NYC_2019.csv")
```

```python
listings
```

```python
sn.countplot(x= 'neighbourhood_group', data = listings)
plt.show()
```

```python
sn.barplot(x= 'neighbourhood_group', y = 'price', data = listings)
plt.show()
```

The wicks that are above the graphs are confidence intervals

```python
sn.barplot(x= 'neighbourhood_group', y = 'price', data = listings, ci = False)
plt.show()
```

Quantative Data Visualisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np

```

```python
listings
```

```python
plt.hist(listings['price'])
plt.xlabel('price(in US Dollar)')
plt.show()
```

Note that above the prices are grouped into contigous intervals called bins

```python
plt.hist(listings['price'], bins = np.arange(0,1100,40))
plt.xlabel('price(in US Dollar)')
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings ['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.show()
```

Restricting the outliers on the x axis

```python
plt.scatter(x = listings['price'], y = listings ['number_of_reviews'])
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.xlim(0,1100)
plt.show()
```

```python
plt.scatter(x = listings['price'], y = listings ['number_of_reviews'], s = 5)
plt.xlabel('price')
plt.ylabel('number_of_reviews')
plt.xlim(0,1100)
plt.show()
```
